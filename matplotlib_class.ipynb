```python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
x = np.linspace(0,5,11)
y = x**2
```

```python
x
```

```python
y
```

```python
plt.plot(x,y)
plt.xlabel('X- axis')
plt.ylabel('Y- axis')
plt.title('Title')
```

```python
plt.subplot(1,2,1)
plt.plot(x,y, 'r')

plt.subplot(1,2,2)
plt.plot(y,x,'g')
```

```python
fig = plt.figure()

axes = fig.add_axes([0.1,0.1,0.8,0.8])

axes.plot(x,y)
axes.set_xlabel('X-label')
axes.set_ylabel('Y-label')
axes.set_title('Set Title')
```

```python
fig = plt.figure()

axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.2,0.5,0.3,0.4])

axes1.plot(x,y)
axes2.plot(y,x)
```

```python
fig,axes = plt.subplots(nrows = 1, ncols=2)
plt.tight_layout()
for current_ax in axes:
    current_ax.plot(x,y)
```

```python
fig = plt.figure(figsize = (3,2), dpi=100)

ax = fig.add_axes([0,0,1,1])
ax.plot(x,x**2, label= 'Square')
ax.plot(x,x**3, label='Cube')
ax.legend()
```
